```python
# Instalasi pustaka yang diperlukan
!pip install opencv-python-headless numpy matplotlib

# Import pustaka
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca file gambar secara langsung
filename = "/content/stadion_old_trafford_manchester_united.jpg"  # Ganti path jika file berada di lokasi lain
image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Periksa apakah gambar berhasil dibaca
if image is None:
    print("Gambar tidak ditemukan atau format tidak valid.")
else:
    # Tampilkan gambar asli
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.show()

# 2. Preprocessing
# Mengubah ukuran gambar agar seragam
resized_image = cv2.resize(image, (512, 512))

# Mengurangi noise menggunakan median filter
denoised_image = cv2.medianBlur(resized_image, 5)

# Tampilkan hasil preprocessing
plt.imshow(denoised_image, cmap='gray')
plt.title('Preprocessed Image')
plt.axis('off')
plt.show()

# 3. Segmentasi
# Konversi gambar ke binary menggunakan Otsu's thresholding
_, binary_image = cv2.threshold(denoised_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Operasi morfologi untuk memperbaiki segmentasi
kernel = np.ones((5, 5), np.uint8)
binary_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)

# Tampilkan hasil segmentasi
plt.imshow(binary_image, cmap='gray')
plt.title('Segmented Image')
plt.axis('off')
plt.show()

# 4. Analisis Gelembung
# Temukan kontur gelembung
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Analisis setiap gelembung untuk menghitung diameter
diameters = []
for contour in contours:
    area = cv2.contourArea(contour)
    if area > 10:  # Abaikan gelembung kecil
        # Hitung diameter ekivalen
        equivalent_diameter = np.sqrt(4 * area / np.pi)
        diameters.append(equivalent_diameter)

# Tampilkan hasil analisis
if diameters:
    print(f"Jumlah gelembung yang terdeteksi: {len(diameters)}")
    print(f"Diameter rata-rata: {np.mean(diameters):.2f} piksel")

    # Plot histogram distribusi diameter
    plt.hist(diameters, bins=10, edgecolor='black')
    plt.title('Distribusi Diameter Gelembung')
    plt.xlabel('Diameter (piksel)')
    plt.ylabel('Frekuensi')
    plt.show()
else:
    print("Tidak ada gelembung yang terdeteksi.")
```
